Install libraries

Step 2 — Load Dataset

```python
import pandas as pd

df = pd.read_csv("Sales.csv")

df.head()
```

Step 3 — Dataset Information

```python
df.info()
```

```python
df.describe()
```

Step 4 — Check Missing Values

```python
df.isnull().sum()
```

Step 5 — Remove Unwanted Columns

```python
df = df.drop(['Day','Month','Year','Age_Group'], axis=1)

df.head()
```

##Why We Removed Them
| Column    | Reason                           |
| --------- | -------------------------------- |
| Day       | Already available in Date        |
| Month     | Can be extracted from Date       |
| Year      | Can be extracted from Date       |
| Age_Group | Can be created from Customer_Age |


Step 6 — Convert Date Column

```python
df['Date'] = pd.to_datetime(df['Date'])

df.info()
```

```python
#Why?
#Because datetime format helps:

#monthly analysis
#yearly trends
#time-series charts
```

Step 7 — Check Duplicate Rows

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

Step 8 — Basic Business Questions

```python
#Total Revenue
total_revenue = df['Revenue'].sum()

print("Total Revenue:", total_revenue)
```

```python
#Total Profit
total_profit = df['Profit'].sum()

print("Total Profit:", total_profit)
```

```python
#Top Selling Product
top_product = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)

print(top_product.head())
```

```python
#Country Wise Sales
country_sales = df.groupby('Country')['Revenue'].sum().sort_values(ascending=False)

print(country_sales)
```

Step 9 — First Visualization

```python
#Monthly Sales Trend
import matplotlib.pyplot as plt

monthly_sales = df.groupby(df['Date'].dt.month)['Revenue'].sum()

monthly_sales.plot(kind='line', figsize=(10,5))

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)

plt.show()
```

```python
#Top 10 Products by Revenue
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar', figsize=(12,6))

plt.title("Top 10 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.xticks(rotation=45)

plt.show()
```

```python
#Country Wise Revenue
country_sales = df.groupby('Country')['Revenue'].sum().sort_values(ascending=False)

country_sales.plot(kind='bar', figsize=(8,5))

plt.title("Country Wise Revenue")
plt.xlabel("Country")
plt.ylabel("Revenue")

plt.show()
```

```python
#Gender Wise Sales
gender_sales = df.groupby('Customer_Gender')['Revenue'].sum()

gender_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))

plt.title("Gender Wise Sales")

plt.ylabel('')

plt.show()
```

```python
#Product Category Analysis
category_sales = df.groupby('Product_Category')['Revenue'].sum()

category_sales.plot(kind='bar', figsize=(8,5))

plt.title("Product Category Revenue")
plt.xlabel("Category")
plt.ylabel("Revenue")

plt.show()
```

```python
#Profit vs Revenue Scatter Plot
plt.figure(figsize=(8,5))

plt.scatter(df['Revenue'], df['Profit'])

plt.title("Revenue vs Profit")
plt.xlabel("Revenue")
plt.ylabel("Profit")

plt.show()
```

```python
#Correlation Heatmap
import seaborn as sns

plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
#KPI Cards
total_revenue = df['Revenue'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order_Quantity'].sum()
average_revenue = df['Revenue'].mean()

print("Total Revenue :", total_revenue)
print("Total Profit :", total_profit)
print("Total Orders :", total_orders)
print("Average Revenue :", average_revenue)
```

```python
#Best Selling Product Category
best_category = df.groupby('Product_Category')['Revenue'].sum().sort_values(ascending=False)

print(best_category)
```

```python
#Age Group Analysis
age_sales = df.groupby('Customer_Age')['Revenue'].sum()

plt.figure(figsize=(12,5))

age_sales.plot()

plt.title("Customer Age vs Revenue")
plt.xlabel("Age")
plt.ylabel("Revenue")

plt.show()
```

```python
#Save Cleaned Dataset
df.to_csv("Cleaned_Sales_Data.csv", index=False)
```
